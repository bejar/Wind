# MLP Dir vs MLP MIMO

```python
from scipy.stats import ks_2samp, ttest_rel,f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
query = {"experiment": "MLPDir","status":"done"}
results = DBResults(conn=mongolocaltest)

results.retrieve_results_dataframe(query)
```

```python
a = results.exp_df.groupby(['site'],as_index=False).sum()
len(a)
```

```python
query = {"experiment": "MLP_s2s_2","status":"done"}
results = DBResults(conn=mongolocaltest)

results.retrieve_results_dataframe(query)
```

```python
b = results.exp_df.groupby(['site'],as_index=False).sum()
c = b.iloc[a['site']]
len(c)
```

```python
plt.figure(figsize=(8,6))
sb.distplot(a['test'],label='MLP dir')
sb.distplot(c['test'],label='MLP s2s')
plt.legend(labels=['MLP dir', 'MLP s2s'], title='Dir vs S2S')
```

```python
ks_2samp(a['test'], c['test'])
```

```python
ttest_rel(a['test'], c['test'])
```

```python
f, p = f_oneway(*[a['test'], c['test']])
f
p
mc = MultiComparison(np.hstack([a['test'], c['test']]), (['MLPDir'] * len(a)) + (['MLPS2S'] * len(a)) )
result = mc.tukeyhsd()
print(result)
```
