# Fitting CNN s2s

## Fit1

```python
query={'experiment':'cnns2sfit'}

results = DBResults(conn=mongolocaltest)
#results = DBResults()

results.retrieve_results_dataframe(query, arch=['filters','strides','kernel_size','drop','full','fulldrop'], data=['lag'])
```

```python
a=results.exp_df.groupby(['site','filters','strides','kernel_size','drop','full','fulldrop','lag'],as_index=False).sum()
a.drop(columns=['hour', 'site'], inplace=True)
```

```python
a[(a['lag']=='12') & (a['filters']=='[64]') &  (a['kernel_size']=='[5]')].boxplot(by=['filters','strides','kernel_size','drop','full','fulldrop','lag'],figsize=(20,10), rot=90)
```

```python
b =a.groupby(by=['filters','strides','kernel_size','drop','full','fulldrop','lag'],as_index=False).mean()
b[(b['lag']=='18') & (b['filters']=='[64]')].style.highlight_max()
```

```python
sns.kdeplot(b['test'],b['val'])
```

Best model CNN(64/1/5,0.1) FC(64,0) L(18)
